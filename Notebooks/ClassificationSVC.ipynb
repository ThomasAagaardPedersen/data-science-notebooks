```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Define the column titles
column_titles = [
    "Compactness", "Circularity", "Distance_Circularity", "Radius_Ratio",
    "Pr_Axis_Aspect_Ratio", "Max_Length_Aspect_Ratio", "Scatter_Ratio", "Elongatedness",
    "Pr_Axis_Rectangularity", "Max_Length_Rectangularity", "Scaled_Variance_Along_Major_Axis",
    "Scaled_Variance_Along_Minor_Axis", "Scaled_Radius_Of_Gyration",
    "Skewness_About_Major_Axis", "Skewness_About_Minor_Axis",
    "Kurtosis_About_Minor_Axis", "Kurtosis_About_Major_Axis", "Hollows_Ratio",
    "Classes"
]
```

```python
# Load the CSV file
# Open each .dat file and save them as .csv 
df1 = pd.read_csv('../datasets/xaa.csv', header=None, delim_whitespace=True)
df2 = pd.read_csv('../datasets/xab.csv', header=None, delim_whitespace=True)
df3 = pd.read_csv('../datasets/xac.csv', header=None, delim_whitespace=True)
df4 = pd.read_csv('../datasets/xad.csv', header=None, delim_whitespace=True)
df5 = pd.read_csv('../datasets/xae.csv', header=None, delim_whitespace=True)
df6 = pd.read_csv('../datasets/xaf.csv', header=None, delim_whitespace=True)
df7 = pd.read_csv('../datasets/xag.csv', header=None, delim_whitespace=True)
df8 = pd.read_csv('../datasets/xah.csv', header=None, delim_whitespace=True)
df9 = pd.read_csv('../datasets/xai.csv', header=None, delim_whitespace=True)

vehicle = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9], ignore_index=True)


## List of file paths
#files = ['xaa.csv', 'xab.csv', 'xac.csv', 'xad.csv', 'xae.csv', 'xaf.csv', 'xag.csv', 'xah.csv', 'xai.csv']

# Use list comprehension to read all CSV files into a list of DataFrames
#dfs = [pd.read_csv(f'../datasets/{file}', header=None, delim_whitespace=True) for file in files]

#df_all = pd.concat(dfs, ignore_index=True)
```

```python
# Assign the column titles
vehicle.columns = column_titles
```

```python
vehicle.head()
```

```python
corr = vehicle.corr()
corr
```

```python
vehicle.drop([
    'Pr_Axis_Aspect_Ratio', "Max_Length_Aspect_Ratio", "Pr_Axis_Rectangularity",
    "Max_Length_Rectangularity", "Scaled_Variance_Along_Minor_Axis",
    "Scaled_Radius_Of_Gyration", "Elongatedness", "Skewness_About_Minor_Axis",
    "Skewness_About_Major_Axis"], axis=1, inplace=True)
```

```python
vehicle.head()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

vehicle['Classes'] = le.fit_transform(vehicle['Classes'].astype(str))

vehicle.head()
```

```python
vehicle['Classes'].unique()
```

```python
x = vehicle.drop('Classes', axis = 1)
y = vehicle['Classes']
```

```python
x = preprocessing.scale(x)

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size = 0.2,
                                                    random_state = 1)
```

```python
from sklearn.svm import LinearSVC

svc_clf = LinearSVC()

svc_clf.fit(x_train, y_train)
```

```python
accuracy = svc_clf.score(x_test, y_test)
print(accuracy)
```

```python
import numpy as np
from sklearn.preprocessing import Normalizer
```

```python
vehicle.head()
```

```python
x_data = vehicle.drop('Classes', axis = 1)
y_data = vehicle['Classes']
```

```python
x_data.head()
```

```python
array = x_data.values #Convert to numpy array
```

```python
array
```

```python
scaler = Normalizer().fit(array)
```

```python
scaler.get_params
```

```python
normalized_array = scaler.transform(array)
```

```python
np.set_printoptions(precision = 3)
print(normalized_array[0:5,:])
```

```python
normalized_df = round(pd.DataFrame(data = normalized_array), 2)
```

```python
column_names = [
    'Compactness',
    'Circularity',
    'Distance_Circularity',
    'Radius_Ratio',
    'Scatter_Ratio',
    'Scaled_Variance_Along_Major_Axis',
    'Kurtosis_About_Minor_Axis',
    'Kurtosis_About_Major_Axis',
    'Hollows_Ratio'
]
```

```python
normalized_df.columns = column_names
```

```python
normalized_data = pd.concat([normalized_df, y_data], axis = 1)

normalized_data.head()
```

```python
x = normalized_data.drop('Classes', axis = 1)
y = normalized_data['Classes']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size = 0.2,
                                                    random_state = 1)
```

```python
svc_clf = LinearSVC()

svc_clf.fit(x_train, y_train)
```

```python
accuracy = svc_clf.score(x_test, y_test)
print(accuracy)
```

```python
pca = PCA(n_components = 7,
          whiten = True)
x_reduced = pca.fit_transform(x)
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
plt.plot(pca.explained_variance_ratio_)

plt.xlabel('Dimension')
plt.ylabel('Explained Various Ratio')

plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_reduced, y,
                                                    test_size = 0.2,
                                                    random_state = 1)
```

```python
svc_clf_pca = LinearSVC(random_state = 5,
                        penalty = 'l1',
                        dual = False,
                        tol = 1e-3,
                        C = 1)
svc_clf_pca.fit(x_train, y_train)
```

```python
accuracy = svc_clf_pca.score(x_test, y_test)
print(accuracy)
```
