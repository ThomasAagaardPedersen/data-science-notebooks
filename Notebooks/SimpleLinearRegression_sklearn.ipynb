```python
#Splitting data set into training and testing
#Shuffles the data sets to avoid biasis (sample function, 1 = 100% of the data)
house_data = house_data.sample(frac = 1).reset_index(drop = True)
```

```python
x_data = house_data['sqft_living']
y_data = house_data ['price']
```

```python
x_data = np.array(x_data)
y_data = np.array(y_data)
```

```python
#x_data_amount = x_data.shape #outputs 21613
# Get the number of data points
x_data_amount = x_data.shape[0]  # Extract integer value (21613)
y_data_amount = y_data.shape[0]  # Ensure we get an integer for y_data as well

# Compute split index
x_data_split = int(x_data_amount * 0.2)
y_data_split = int(y_data_amount * 0.2)  # Should be the same as x_data_split if x and y match
```

```python
#Reshape the data into 2 dimensional array
x_data = x_data.reshape(-1, 1)
y_data = y_data.reshape(-1, 1)
```

```python
#80:20 Train:Test split
train_x_data = x_data[:x_data_split] //-4322
test_x_data = x_data[x_data_split:] //-4322

train_y_data = y_data[:y_data_split] //-4322
test_y_data = y_data[y_data_split:] //-4322
```
