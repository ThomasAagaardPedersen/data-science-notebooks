```python
# import the libraries required
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
```

```python
# unit basis vectors (typical Cartesian axis basis vectors)
ihat = np.array([1, 0])
jhat = np.array([0, 1])

# the vector v
v = np.array([2, 2])

# now we define a scalar s, such that s multiplied by the magnitude of the vector v
# we do this operation just to create a unit vector in the same direction as the vector v
# note that the magnitude of the vector v is ~ 2.82843 or (2*(root 2))
length = la.norm(v)
s = 1 / la.norm(v)

# we can confirm this by multiplying by la.norm, thus the proof s * la.norm(v) = char 's' * ln.norm(v)
check = s * la.norm(v)

# now we calculate the unit vector (scale) based on the scalar computed value:
vunit = v * s

# dot product of v_unit with unit basis vector i_hat
# this dot product is +ve when both vectors are facing the same direction
# not using np.dot as numpy does recommend '@' when the operators are vectors
dp = vunit @ i_hat

# configure plot values, formatting etc.
# note that projection of v_unit on i_hat, j_hat is v_unit @ i_hat, v_unit @ j_hat
# demonstrating that the dot product is the cosine between two vectors
plt.plot([0, v[0]], [0, v[1]], 'b:', label='v', linewidth=3)
plt.plot([0, vunit[0]], [0, vunit[1]], 'r', label='v_unit', linewidth=3)
plt.plot([0, 1], [0, 0], 'r:', label='i_hat', linewidth=4)
plt.plot([0, 0], [0, 1], 'r:', label='j_hat', linewidth=4)
plt.plot([0, vunit[0]], [0, 0], 'k', label='prj on x', linewidth=2)
plt.plot([0, 0], [0, vunit[1]], 'k', label='prj on y', linewidth=2)

# generate the plot with axis, grid, legend
print("Confirm 1/la.norm * la.norm =", check)
print("Length of vector v:", length)
print("Vector v_unit in same direction as v:", vunit)
print("Dot product v_unit with i_hat +ve when facing same direction:", dp)
plt.axis('square')
plt.axis((-1, 3, -1, 3))
plt.grid()
plt.legend()
plt.show()
```
